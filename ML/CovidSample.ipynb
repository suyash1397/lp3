```python
import pandas as pd
df=pd.read_csv("Iris.csv")
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x='SepalLengthCm',y='SepalWidthCm',data=df)
```

```python
labels = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
sizes = [30, 45, 15, 10]
colors = ['lightcoral', 'lightgreen', 'lightblue', 'lightsalmon']

# Create a pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Fruit Distribution')

# Display the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
data = df

# Create a violin plot
sns.violinplot(x="PetalWidthCm", y="PetalLengthCm", data=data)

# Add labels and a title
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.title("Violin Plot of Total Bill by Day")
plt.show()
```

```python
data = sns.load_dataset("iris")
correlation_matrix = data.corr()
plt.figure(figsize=(8, 6))   ##LabelEncoding
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
