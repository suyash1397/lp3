```python
x=3
lr =0.01
pricision = 0.000001
prev_step_size = 1
iters =0
max_iters =10000
gf = lambda x: (x+3)**2
Ploting_list = []
```

```python
while pricision<prev_step_size and iters <max_iters:
    prev =x
    x = x-lr*gf(prev)
    prev_step_size = abs(x-prev)
    iters+=1
    Ploting_list.append(x)
    print("Items: ",iters,"Value",x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(Ploting_list)
```
